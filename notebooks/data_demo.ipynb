```python
from xray_project.dataset import ChestXrayDataset
from torchvision import transforms
```

```python
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])
```

```python
dataset = ChestXrayDataset(
    csv_file="../data/sample_labels.csv",
    image_dir="../data/sample_images",
    transform=transform
)
```

```python
image, disease_labels, metadata_label = dataset[0]

print("Image shape:", image.shape)
print("Disease labels:", disease_labels)
print("Metadata label:", metadata_label)
```

```python
import matplotlib.pyplot as plt

plt.imshow(image.permute(1,2,0))
plt.axis("off")
plt.show()
```
